```python
"""
This cell contains a functions the function `build_Q` that returns the stochstic matrix coresponding to Q in the description above. 

Additonal functions that may be useful are also included.
The function `letters_to_integers` converts a string to a list of integers based on a dictionary.
The function `integers_to_letters` converts a list of integers to a string based on a dictionary.
The function `integers_to_binary` converts a list of integers (0:(2^k-1)) to a binary string of length k * len(lo_integers).
The function `binary_to_integers` converts a binary string of length k * len(s) to a list of integers.
The function `entropy` calculates the entropy of a discrete distribution.

"""
import numpy as np
from matplotlib import pyplot as plt


YOUR_ID_NUMBER = 42 # modify this to your ID number
np.random.seed(YOUR_ID_NUMBER)

def build_Q():
    """
    Returns:
    Q: a 13x13 numpy array representing the transition matrix for the Markov chain.
    """
    
    Q1 = np.array([[1, 1, 2], [1, 1, 1], [1, 1, 1]])
    Q = np.zeros((13, 13))
    Q[:3, :3] = Q1
    Q[2:5, 2:5] = Q1
    Q[4:7, 4:7] = Q1
    Q[6:9, 6:9] = Q1
    Q[8:11, 8:11] = Q1
    Q[10:13, 10:13] = Q1
    Q[0, 0] = Q[1,1] = 9
    Q[2, 2] = Q[3,3] = 8
    Q[4, 4] = Q[5,5] = 7
    Q[6, 6] = Q[7,7] = 6
    Q[8, 8] = Q[9,9] = 5
    Q[10, 10] = Q[11,11] = 4
    Q[12, 12] = 3
    Q[0, 2] = 1
    Q[10, 12] = 1
    return Q / Q.sum(axis=1, keepdims=True)

Xcal = list('abcdefghijklm')


def letters_to_integers(s, dictionary):
    """
    Convert a string to a list of integers based on a dictionary.
    """
    return [dictionary.index(c) for c in s]

def integers_to_letters(lo_integers, dictionary):
    """
    Convert a list of integers to a string based on a dictionary.
    """
    return [dictionary[i] for i in lo_integers]

def integers_to_binary(lo_integers, k=4):
    """
    Convert a list of integers (0:(2^k-1)) to a binary string of length k * len(lo_integers).
    """
    return "".join([np.binary_repr(i, width=k) for i in lo_integers])

def binary_to_integers(s, k=4):
    """
    Convert a binary string of length k * len(s) to a list of integers.
    """
    return [int(s[i:i+k], 2) for i in range(0, len(s), k)]
```

```python
# 2. What is the stationary distribution? 
# (write a code that evaluate this distribution and plot it; make sure to label the states correctly with letters).
def get_stationary_distribuition(Q):
  eigen_values, eigen_vectors = np.linalg.eig(Q.T)
  eigen_vec_to_val_one = eigen_vectors[:,np.isclose(eigen_values, 1)]

  stationary_dist = eigen_vec_to_val_one[:, 0]
  #ensure sums to one
  stationary_dist = stationary_dist / stationary_dist.sum()
  stationary_dist = stationary_dist.real
  return stationary_dist

def plot_stationary_distribuition(stationary_dist):
  labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm']
  plt.bar(labels, stationary_dist.real)
  plt.xlabel("states")
  plt.ylabel("probabilities")
  plt.show()

get_stationary_distribuition(Q)
plot_stationary_distribuition(get_stationary_distribuition(Q))
```

```python
# some huffman functions
# code to build huffman code compressor

# node class
class node:
  def __init__(self, symbol, prob):
    self.symbol = symbol
    self.prob = prob
    self.left = None
    self.right = None

def create_node_list(label_prob_dict):
  return [node(symbol, prob) for symbol, prob in label_prob_dict.items()]


# changed to smallest (change name)
def pop_largest_prob(my_node_queue):
  smallest_prob_index = 0
  for index, n in enumerate(my_node_queue):
    if n.prob < my_node_queue[smallest_prob_index].prob:
      smallest_prob_index = index
  return smallest_prob_index, my_node_queue[smallest_prob_index] 

def build_huffman_tree(symbol_prob_dict):
    node_list = create_node_list(symbol_prob_dict) 
    while len(node_list) > 1:
      i, left_node = pop_largest_prob(node_list)
      node_list.pop(i)

      i, right_node = pop_largest_prob(node_list)
      node_list.pop(i)     

      merged_node = node(None, left_node.prob + right_node.prob) 
      merged_node.left = left_node
      merged_node.right = right_node

      node_list.append(merged_node)
    return node_list[0]



# 1 find impirical distribution
def find_empirical_dist(Q, n, symbols):
  sample = sample_from_chain(Q, n, 13, symbols)
  return np.unique(sample, return_counts = True)[1] / np.unique(sample, return_counts = True)[1].sum(), sample


def make_dict(symbols, empirical_dist):
  return {symbol: prob for symbol, prob in zip(symbols, empirical_dist)}


def get_code(node, path_str, huffman_code):
  if node:
    if node.symbol is not None:
        huffman_code[node.symbol] = path_str
    else:
        get_code(node.left, path_str + "0", huffman_code)
        get_code(node.right, path_str + "1", huffman_code)
    return huffman_code

```

```python

def move_one_up(code_dict, symbol):
    inputed_value = code_dict[symbol]
    for k, v in  code_dict.items():
        if v < inputed_value:
            code_dict[k] = v + 1
    code_dict[symbol] = 0
    return code_dict

def inverse_dict(code_dict):
    revered_dict = {v : k for k, v in code_dict.items()}
    return revered_dict

def move_to_front_transform(input, max_int = 13):
    """
    Move-to-front transform of a list of integers. The list of integers is assumed to consist of the integers 0, 1, ..., 12 in some order.

    Args:
    input: a list of integers
    max_int: the maximum integer in the list
    """

    # YOUR CODE HERE
    code_dict = {symbol: symbol for symbol in range(max_int)}
    output = []
    for symbol in input:
        output.append(code_dict[symbol])
        code_dict = move_one_up(code_dict, symbol)
    return output



def inverse_move_to_front_transform(input, max_int = 13):
    """
    Inverse move-to-front transform of a list of integers. The list of integers is assumed to consist of the integers 0, 1, ..., 12 in some order.

    Args:
    input: a list of integers
    max_int: the maximum integer in the list

    """
   
    # YOUR CODE HERE
    code_dict = {symbol: symbol for symbol in range(max_int)}
    output = []
    for symbol in input:
        revered_dict = inverse_dict(code_dict)
        output.append(revered_dict[symbol])
        code_dict = move_one_up(code_dict, revered_dict[symbol])
        # print(code_dict)
        # print(output)
    return output




# test
i = [0, 1, 0, 0, 5, 2, 0, 6, 0,8,9,9,3,2,0,0,0,0,0,1,1,1,1]
c = move_to_front_transform(i, max_int = 13)
print(c)
b = inverse_move_to_front_transform(c, 13)
print(i == b)
print(b)
```

```python
YOUR_ID_NUMBER = 806614 # modify this to your ID number
np.random.seed(YOUR_ID_NUMBER)

def send_through_channel(x, p):
    """
    Simulate a binary symmetric channel with crossover probability p
    """
    z = np.random.rand() < p
    return x ^ z # XOR

def sample_message(n, R):
    """
    Draw a random message out of ceil(2^(nR)) possible messages
    """
    return np.random.randint(np.ceil(2**(R*n)))

def build_random_codebook(n, R):
    """
    Generate a random codebook of 2^(nR) codewords
    """
    return np.random.randint(2, size=(int(np.ceil(2**(R*n))), n))

def encode_message(W, codebook):
    """
    Encode a message W using a codebook
    """
    return codebook[W]

def decode_message(y, codebook, epsilon, p):
    """
    Decode a received message y using the codebook using joint typicality
    In this version, if we find more than one typical codeword we return 
    the index of the first one. We return None if no typical codeword is found.

    Args:
    y: received message
    codebook: the codebook (2D array of size 2^(nR) x n)
    epsilon: joint typicality threshold
    p: channel crossover probability
    """

    # YOUR_CODE_HERE

    # from Q5 to check if (x_hat, y) is typical we can just check if z is typical
    n = len(y)
    found_index = []
    H = -1 * (p * np.log2(p) + (1-p) * np.log2(1-p))
    dict = typ_dict(n, epsilon, p=p, H=H)
    for index, codeword in enumerate(codebook):
        num_flips = np.sum(codeword != y)
        if dict[num_flips] == True:
            found_index.append(index)



    if len(found_index) == 1:
        return found_index[0]
    elif len(found_index) > 1:
        return -1
    else:
        return None
    
 
def simulate_transmission(w, codebook, epsilon, p, verbose=False):
    """
    Simulate the transmission of a message through the channel

    w: message index
    codebook: the codebook (2D array of size 2^(nR) x n)
    epsilon: joint typicality threshold
    p: channel crossover probability

    Returns the decoded message
    """
    x = encode_message(w, codebook)
    y = np.zeros_like(x)
    n = len(x)
    for i in range(n):  # sending n transmissions through the channel
        y[i] = send_through_channel(x[i], p)
    w_hat = decode_message(y, codebook, epsilon, p)

    if verbose:
        print("Message: ", w)
        print("Encoded message:  ", x)
        print("Received message: ", y)
        print("Decoded message: ", w_hat)
    return w_hat

def check_transmission(N, n, R, epsilon, p):
    """
    Check the transmission of N random messages through the channel.

    N: number of messages to simulate
    n: number of channel uses
    R: code rate
    p: channel crossover probability
    epsilon: joint typicality threshold

    Returns a dictionary with the number of non-typical errors, the number of more than one typical errors.
    """

    codebook = build_random_codebook(n, R)  # generate a random codebook
    
    num_of_non_typical_error = 0      # update this number in the loop
    num_of_more_other_error = 0       # update this number in the loop

    for _ in range(N):
        w = sample_message(n, R)
        w_hat = simulate_transmission(w, codebook, epsilon, p)

        # YOUR_CODE_HERE
        if w_hat == -1:
            num_of_more_other_error += 1
        if w_hat is None:
            num_of_non_typical_error += 1 

    return dict(num_of_non_typical_errors=num_of_non_typical_error,
            num_of_more_other_decoding_errors=num_of_more_other_error)
```
